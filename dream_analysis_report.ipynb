```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\Asus\\Downloads\\Dream_Sentiment_Analysis_01.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Dream_Sentiment_Score'],inplace=True)
```

```python
df
```

```python
df.index
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Caffeine consumption'].fillna(df['Caffeine consumption'].median(),inplace=True)
df['Alcohol consumption'].fillna(df['Alcohol consumption'].median(),inplace=True)
df['Deep sleep percentage'].fillna(df['Deep sleep percentage'].median(),inplace=True)
```

```python
df['Awakenings'].fillna(df['Awakenings'].mean(),inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("cleaned_dream_data.csv", index=False)
```

```python
df.isnull().sum().any()
```

```python
df
```

```python
X=df.drop(columns=['Dream_Sentiment_Score'])
```

```python
X
```

```python
y=df['Dream_Sentiment_Score']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.32,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor
treeclss=DecisionTreeRegressor()
```

```python
treeclss.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(20,30))
tree.plot_tree(treeclss,filled=True)
```

pre-pruning


```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [3, 5, 10, 15, 20],  # Increase depth for better learning
    'min_samples_split': [2, 5, 10],  # Avoid overfitting
    'min_samples_leaf': [1, 3, 5, 10],  # Ensure meaningful splits
    'ccp_alpha': [0.001, 0.002, 0.003, 0.005]  # Pruning parameter
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(treeclss,param_grid=param,cv=5,scoring='r2',n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
best_tree = DecisionTreeRegressor(
    criterion='absolute_error',   # From GridSearchCV
    max_depth=20,
    splitter='best',
    min_samples_leaf=1,
    min_samples_split=2,
    ccp_alpha=0.001, 
    random_state=42  # Ensure reproducibility
)

# # Train with a deeper tree
# best_tree = DecisionTreeRegressor(
#     criterion="squared_error",
#     max_depth=3,  # Increase depth
#     min_samples_split=5,  # Avoid overfitting
#     min_samples_leaf=3,  # Ensure meaningful splits
#     random_state=42
# )
```

```python
best_tree.fit(X_train, y_train)
```

```python

plt.figure(figsize=(10,10))
tree.plot_tree(best_tree,filled=True)
```

```python
y_pred = best_tree.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred)  # R² Score (higher is better)
mae = mean_absolute_error(y_test, y_pred)  # MAE (lower is better)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # RMSE (lower is better)
```

```python
print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Evaluate on training data
y_train_pred = best_tree.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)

# Evaluate on test data
y_test_pred = best_tree.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)

print(f"Training R² Score: {r2_train:.4f}")
print(f"Testing R² Score: {r2_test:.4f}")
```

```python
import pickle

with open("dream_sentiment_model.pkl", "wb") as model_file:
    pickle.dump(best_tree, model_file)
print("Model saved successfully with pickle!")
```
